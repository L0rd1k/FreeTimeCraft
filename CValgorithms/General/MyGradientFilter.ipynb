```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('../Images/Cropped_Runway/runway.png')
plt.figure(figsize=(8,4))
plt.imshow(image)

def showImageRGB(image):
    plt.axis('off')
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

kernelSize = 3
aplhaValue = 2
kernel = cv2.getGaussianKernel(kernelSize, 0)
print("Gaussian kernel: ",kernel)
plt.imshow(kernel)
```

```python
kernel = -aplhaValue * kernel @ kernel.T
print("Gaussian kernel: ",kernel)
plt.imshow(kernel)
```

```python
kernel[kernelSize//2, kernelSize//2] += 1 + aplhaValue
print(kernel.shape, kernel.dtype, kernel.sum())
plt.imshow(kernel)
```

```python
filtered = cv2.filter2D(image, -1, kernel)
print(filtered.shape, filtered.dtype, filtered.sum())
plt.imshow(cv2.cvtColor(filtered,cv2.COLOR_BGR2RGB))
```
