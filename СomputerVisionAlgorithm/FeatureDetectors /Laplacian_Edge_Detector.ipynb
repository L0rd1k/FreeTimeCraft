```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
plt.figure(figsize=(100,100))

image = cv2.imread('/home/ilya/AnacondaProjects/WolrdAroundPython/Images/Cropped_Runway_Images/00104.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.subplot(3,3,1)
plt.imshow(image)
print(image.shape)

gray_images = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
image  = cv2.GaussianBlur(gray_images,(3,3),0)
plt.subplot(3,3,2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_GRAY2RGB))
print(image.shape)

laplacian = np.array(cv2.Laplacian(image, cv2.CV_64F), dtype = int)
plt.subplot(3,3,3)
plt.imshow(laplacian, cmap = 'binary')
plt.show()
```
