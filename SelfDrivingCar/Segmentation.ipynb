```python
import cv2
import numpy as np

image = cv2.imread('./images/Paper.jpg')

def imageSortingContours(image):
    blank_image = np.zeros((image.shape[0],image.shape[1],3))
    original_image = image
    gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    edged = cv2.Canny(gray,50,200)
    cv2.imshow('Canny edges',edged)
    #find contours
    contours,hierarchy = cv2.findContours(edged.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
    print("Number of contours: ",len(contours))
    #draw them
    cv2.drawContours(blank_image,contours,-1,(0,255,0),1)
    cv2.imshow("Blank contours:",blank_image)
    cv2.drawContours(image,contours,-1,(0,255,0),1)
    cv2.imshow("All contours:",image)
    return contours


def get_contour_areas(contours):
    all_areas = []
    for cnt in contours:
        area = cv2.contourArea(cnt)
        all_areas.append(area)
    return all_areas

def imageStepContours(image,contours):
    original_image = image
#     print(get_contour_areas(contours))
#     sorted_contours = sorted(contours,key=cv2.contourArea,reverse=True)
    sorted_contours = sorted(contours,key=cv2.contourArea,reverse=True)[:3]
#     print(get_contour_areas(sorted_contours))
    for c in sorted_contours:
        cv2.drawContours(original_image,[c],-1,(255,0,0),1)
        cv2.imshow("Contours by area", original_image)
        cv2.waitKey(25)
        
def x_cord_contour(contours):
    if cv2.contourArea(contours) > 10:
        M = cv2.momnets(contours)
        return(int(M['m10']/M['m00']))
    
# def label_contour_center(image,c,i):
#     M = cv2.moments(c)
#     cx = int(M['m10'] / M['m00'])
#     cy = int(M['m01'] / M['m00'])
#     cv2.circle(image,(cx,cy),10,(0,0,255),-1)
#     return image

# def imageCentralPoint(image,contours):
#     original_image = image.copy()
#     for(i,c) in enumerate(contours):
#         orig = label_contour_center(image,c,i)
   
    
# imageSortingContours(image)  
contours = imageSortingContours(image)
# imageCentralPoint(image,contours);
imageStepContours(image,contours)
cv2.waitKey(0)
```
