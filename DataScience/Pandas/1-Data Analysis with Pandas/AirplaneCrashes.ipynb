```python
# преобразование типов
table['Date'] = pd.to_datetime(table['Date']) 
table['Time'] = pd.to_datetime(table['Time'], errors = 'coerce')
# Достаньте из 'Date' - Год, месяц, день недели. 
table['year'] = table['Date'].dt.year
table['month'] = table['Date'].dt.month
table['day'] = table['Date'].dt.day
table['day of week'] = table['Date'].dt.dayofweek # Monday = 0, Sunday = 6,
table['Hours'] = table['Time'].dt.hour

table.head(3)
#Посмотрите сколько инцидентов было в 2002 году: 
table[table['year'] == 2002].shape[0]
#Посмотрите сколько инцидентов было c кол-вом жертв (суммарно на земле и на борту) больше 200: 
table[table['Fatalities'] + table['Ground'] > 200].shape[0]
# Кроме того мы можем производить любые операции со столбцами складывать/умножать/делить/логарифимировать etc.   
table['all_victims'] = table['Fatalities'] + table['Ground']
# Посмотрите сколько инцидентов было c кол-вом жертв (суммарно на земле и на борту) больше 200 в 2001 году:
table[(table['year'] == 2001) & (table['all_victims'] > 200)].shape[0] #head()
# Где произошел инцидент? Какой был самолет? Сколько было жертв?
for i in range(0,table[(table['year'] == 2001) & (table['all_victims'] > 200)].shape[0]):
    print('В 2002 году в', table[(table['year'] == 2001) & (table['all_victims'] > 200)]['Location'].values[i],
    'разбился',table[(table['year'] == 2001) & (table['all_victims'] > 200)]['Type'].values[i],'. ' 
    'Общее колличество жертв',table[(table['year'] == 2001) & (table['all_victims'] > 200)]['Location'].values[i],      
    )
    pass
```

```python
location = table['Location'].str.split(',',expand=True)
location[3] = location[3].fillna(location[2])
location[3] = location[3].fillna(location[1])
location[3] = location[3].fillna(location[0])
location[3].nunique()
#location.head()
```

```python
# объединим таблицы с помощью pd.merge()
table_full = pd.merge(table,location,left_index=True,right_index=True)
table_full.head(3)
# Удалите колонки '0', '1', '2' с помощью .drop() 
# переименуйте колонку '3' в  'Country' (по крайней мере мы надеемся, что там страна)
table_full = table_full.drop([0,1,2], axis=1)
table_full.columns.values[19] = 'Country'
table_full.head(3)
```

```python
# Сделаем признаки: Day_period и Holiday с помощью .apply(lambda x: ...)
def get_day_time(hour):
    return 
    {
        0 <= hour < 6:   'Ночь',
        6 <= hour < 11:  'Утро',
        11 <= hour < 19:  'День',
        19 <= hour < 23:  'Вечер',
        23 <= hour < 25:  'Ночь'
    }[True]
table_full['Day_period'] = table_full['Hours'].fillna(0).apply(lambda x: get_day_time(x))
table_full['Holiday'] = table_full['day of week'].fillna(0).apply(lambda x: 1 if x >=5 else 0)
table_full.head()
```
