```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(palette='deep', style='darkgrid', rc={"figure.figsize": (15, 8)})

import scipy.stats as st

import warnings
warnings.simplefilter('ignore')
```

```python
table = pd.read_csv('data/Airplane_Crashes_and_Fatalities_Since_1908.csv')
#table.head() #or $table.tail()
table['Time'] = table['Time'].str.replace('c:','')
```

```python
#shape позволяет выводить размер датасета.
print(table.shape) #or table.shape[0] #table.shape[1]
```

```python
#сформирует таблицу с типами данных с помощью класса .dtypes:
table_info = pd.DataFrame(table.dtypes,columns=['Type'])
#table_info
```

```python
#Посмотрим уникальные значения Авиаоператоров с помощью.unique() и их кол-во .nunique()
#table['Operator'].unique()
table['Operator'].nunique()
```

```python
# класс info
table.info()
```

```python
# РАБОТА С ДАТОЙ
```

```python
# Ошибки можно проигнорировать параметром errors = 'coerce'
table['Data'] = pd.to_datetime(table['Date'])
table['Time'] = pd.to_datetime(table['Time'],errors = 'coerce')
```

```python
table['years'] = table['Data'].dt.year
table['month'] = table['Data'].dt.month
table['day'] = table['Data'].dt.day
table['day of week'] = table['Data'].dt.dayofweek # 0-Moday and etc.

table['Hours'] = table['Time'].dt.hour
table.head(3)
```

```python
table['all_victims'] = table['Fatalities'] + table['Ground']
table.head()
#loc gets rows (or columns) with particular labels from the index.
#iloc gets rows (or columns) at particular positions in the index (so it only takes integers).
```

```python
#РАБОТА С ТЕКСТОМ И ПРОПУСКАМИ,ОБЪЕДИНЕНИЕ ТАБЛИЦ
```

```python
#Разобьем название через запятую
#fillna - заполнить пропуски
#drop - удалить пропуски
loc = table['Location'].str.split(',',expand=True)
loc[3] = loc[3].fillna(loc[2])
loc[3] = loc[3].fillna(loc[1])
loc[3] = loc[3].fillna(loc[0])
loc[3].nunique()
```

```python
# объединим таблицы с помощью pd.merge()
table_full = pd.merge(table,loc,left_index=True,right_index=True)
table_full.head(2)
```

```python
# Удалите колонки '0', '1', '2' с помощью .drop()
table_full = table_full.drop([0,1,2], axis=1)
# Переименовть колонку "3" в Country
table_full.columns.values[20] = 'Country'
table_full.head(3)
```

```python
def get_day_time(hour):
    return {
          0 <= hour < 6:   'Ночь',
          6 <= hour < 11:  'Утро',
         11 <= hour < 19:  'День',
         19 <= hour < 23:  'Вечер',
         23 <= hour < 25:  'Ночь'
    }[True]

table_full['Day_period'] = table_full['Hours'].fillna(0).apply(lambda x: get_day_time(x))
table_full['Holiday'] = table_full['day of week'].fillna(0).apply(lambda x: 1 if x >= 5 else 0) 
```

```python
# map применяет какую-либо функцию к списку и возвращает результат опять же в виде списка.
```

```python
#СВОДНЫЕ ТАБЛИЦЫ, ГРУППИРОВКА И АГРЕГАЦИЯ ДАННЫХ
table_full.head()
```

```python
#посчитайть сумму жертв по годам - .groupby
#сортировать по убыванию - sort_values(...,ascending=False)
table_full.groupby('years').sum().sort_values('all_victims', ascending=False).head()
```

```python
table.groupby('years').sum()[['all_victims','Ground','Fatalities']].plot(rot=45)
plt.title('Колличество жертв по годам')
plt.show()
```

```python
#Посчитайте кол-во Инцидентов и жертв по годам
crashes_by_year = pd.pivot_table(table_full,index='years',values=['Aboard','Fatalities','Ground','all_victims'], aggfunc='sum')
crashes_by_year['count'] =  pd.pivot_table(table_full, index='years', values=['all_victims'], aggfunc='count').values.T[0]
crashes_by_year.tail()
```

```python
# Сколько жертв на инцидент? 
crashes_by_year['Victims_on_incident'] = crashes_by_year['all_victims']/crashes_by_year['count']
#crashes_by_year.head()
# существует ли корреляция между всеми праметрами? 
crashes_by_year.corr()
```

```python
sns.heatmap(crashes_by_year.corr(), vmin=-1, vmax=1,annot=True, cmap='RdBu',)
plt.show()
```

```python
_, axes = plt.subplots(1, 2, sharey=True, figsize=(16,6))

sns.boxplot(x="Day_period", y="Fatalities", data=table_full[table_full['Fatalities'] < 100],  ax=axes[0]);
sns.violinplot(x="Day_period", y="Fatalities", data=table_full[table_full['Fatalities'] < 100], ax=axes[1], hue='Holiday');
```

```python
fig = plt.figure(figsize=(22, 5))
ax1 = fig.add_subplot(121)
sns.distplot(table_full['Fatalities'].dropna(), fit=st.norm, kde=True,ax=ax1, bins=50)
ax1.set_title('Плотность распределения переменной')
ax2 = fig.add_subplot(122)
prob = st.probplot(table_full['Fatalities'].dropna(), dist=st.norm, plot=ax2)
ax2.set_title('Probabilyty plot')
plt.show()
```
